```python
label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

(unique, counts) = np.unique(y_train, return_counts=True)
```

```python
X_train_scaled=x_train/255 #scaling
X_test_scaled=x_test/255
```

```python
X_train_scaled[0]
```

```python
#Model
num_classes=10

from tensorflow import keras

#layers

model1=keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(256,activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(num_classes,activation='softmax')
])

model1.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['acc'])
```

```python
model1.fit(X_train_scaled,y_train,
            validation_split=0.1,
            epochs=20)
```

```python
y_pred=model1.predict(X_test_scaled)
y_pred_labels = np.argmax(y_pred, axis=1)
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, classification_report

acc = accuracy_score(y_test, y_pred_labels)
print(f"Accuracy: {acc:.4f}")

# F1 Score (macro average for multiclass)-avg of f1 of each class
f1 = f1_score(y_test, y_pred_labels, average='macro')
print(f"F1 Score (macro): {f1:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_labels))
```

# Resnet - Transfer Learning

```python
from tensorflow.keras import Sequential,models,layers
from tensorflow.keras.layers import Dense,Dropout,Flatten
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras import optimizers
```

```python
Convolution_Base=ResNet50(weights='imagenet',include_top=False, input_shape=(256,256,3))
Convolution_Base.summary()
```

```python
model2=models.Sequential()
model2.add(layers.UpSampling2D((2,2)))
model2.add(layers.UpSampling2D((2,2)))
model2.add(layers.UpSampling2D((2,2)))
model2.add(Convolution_Base)
model2.add(layers.Flatten())
model2.add(layers.BatchNormalization())
model2.add(layers.Dense(128,activation='relu'))
model2.add(layers.Dropout(0.5))
model2.add(layers.BatchNormalization())
model2.add(layers.Dense(64,activation='relu'))
model2.add(layers.Dropout(0.3))
model2.add(layers.BatchNormalization())
model2.add(layers.Dense(num_classes,activation='softmax'))

model2.compile(optimizer=optimizers.RMSprop(learning_rate=2e-5),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'],          
             )
```

```python
history=model2.fit(X_train_scaled,y_train,validation_split=0.1,epochs=10)
```

```python
h=history

plt.plot(h.history['loss'],label='training_loss')
plt.plot(h.history['val_loss'],label='validation_loss')
plt.legend()
plt.show()

plt.plot(h.history['acc'],label='training_acc')
plt.plot(h.history['val_acc'],label='validation_accuracy')
plt.legend()
plt.show()
```
